```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer, FunctionTransformer

# Create a highly skewed dataset (exponential)
X = np.random.exponential(scale=2, size=(1000, 1))

# Apply PowerTransformer (Yeo–Johnson)
pt = PowerTransformer(method='yeo-johnson')
X_pt = pt.fit_transform(X)

# Apply FunctionTransformer (log1p)
ft = FunctionTransformer(np.log1p)
X_ft = ft.fit_transform(X)

# Plot before and after transformations
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.hist(X, bins=30, color='gray')
plt.title("Original Data (Skewed)")

plt.subplot(1, 3, 2)
plt.hist(X_pt, bins=30, color='skyblue')
plt.title("PowerTransformer (Yeo–Johnson)")

plt.subplot(1, 3, 3)
plt.hist(X_ft, bins=30, color='lightgreen')
plt.title("FunctionTransformer (log1p)")

plt.tight_layout()
plt.show()
```
